```python
from pycaret.regression import *
import pandas as pd

# Leer tus datos
# Reemplaza 'tu_archivo.csv' con el nombre de tu archivo de datos
data = pd.read_csv("CM01_3_avg_amount_procesado.csv")

data = data.head(int(data.shape[0] / 2))
# Configurar PyCaret para regresión
regression_setup = setup(data, target="avg_amount", n_jobs=4)

# Comparar modelos y seleccionar el mejor
best_model = compare_models()

# Crear un modelo específico con el mejor algoritmo seleccionado
final_model = create_model(best_model)

# Evaluar el modelo en el conjunto de prueba
evaluate_model(final_model)

# Realizar predicciones en un nuevo conjunto de datos
# Reemplaza 'new_data.csv' con el nombre de tu nuevo conjunto de datos
new_data = pd.read_csv("CM01_3_avg_amount_procesado.csv")
new_data = new_data.drop("avg_amount", axis=1)
predictions = predict_model(final_model, data=new_data)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Supongamos que 'y_true' son los valores reales y 'y_pred' son las predicciones
data = pd.read_csv('CM01_3_avg_amount_procesado.csv')
y_true = data['avg_amount']
y_pred = predictions['prediction_label']

# Calcular métricas
mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

# Visualizar resultados
plt.scatter(y_true, y_pred)
plt.xlabel('Valores reales')
plt.ylabel('Predicciones')
plt.title('Gráfico de dispersión entre valores reales y predicciones')
plt.show()

# Imprimir métricas
print(f'MAE: {mae:.2f}')
print(f'MSE: {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'R^2: {r2:.2f}')
```
